```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("turkey_car_market.csv")
```

```python
# Listeler
df
```

```python
# Sütun başlıklarını listeler
df.columns
```

```python
# İsim değiştirir;
df.rename(columns ={"Arac Tip":'Araba_tip'},inplace=True)
```

```python
#ilk "5" satırı sıralar.
df.head(5)
```

```python
#sütün silme kodu
df.drop('Araba_tip',axis=1, inplace=True) 
```

```python
df.head(5)
```

```python
df["Ne_Arabası"]="" #sütun ekleme
```

```python
df.head(5)
```

```python
para_durumu = []
for i in df.Fiyat:
    if i <= 20000:
        para_durumu.append('Gariban')
    elif i > 20000 and i <=120000:
        para_durumu.append('Normal')
    else:
        para_durumu.append('Zengin')
```

```python
df['Ne_Arabası'] = para_durumu
df.head(5)
```

```python
# markaların model yıllarını küçüğe büyükten sıralar;
df.sort_values(['Marka','Model Yıl']) 
```

```python
# Km göre düşük,yüksek;

# Veri setindeki 'Km' sütununun ortalamasını hesaplayın
ortalama_km = sum(df.Km) / len(df.Km)

# 'Km miktar' adında yeni bir sütun oluşturun
# Eğer aracın kilometresi ortalama değerden büyükse 'Yüksek', aksi takdirde 'Düşük' değeri atayın
df["Km miktar"] = ["Yüksek" if km > ortalama_km else "Düşük" for km in df.Km]

# İlk 10 satırda bulunan 'Km miktar' ve 'Km' sütunlarını seçin ve ekrana yazdırın
ilk_20_satir = df.loc[:20, ["Marka","Km miktar", "Km"]]

# Sonuçları görüntüleyin
print(ilk_20_satir)
```

```python
#  Marka sütununun içindeki değerlerin kaç kere tekrarlandığını gösterir
df["Marka"].value_counts()
```

```python
# 'Marka' sütunundaki değer sayılarını içeren bir değişken oluşturun
top10 = df['Marka'].value_counts()[:10]
```

```python
# 'fivethirtyeight' stili kullan
plt.style.use("fivethirtyeight")

# Çubuk grafiği için yeni bir figure oluştur
plt.figure(figsize=(16, 6))

# Çubuk grafiğini çiz
plt.bar(top10.index, top10.values,color="#00A7FF")

# Grafiği göster
plt.show()
```

```python
# 'Renk' sütunundaki değer sayılarını içeren bir değişken oluşturun
top10color = df['Renk'].value_counts()[:10]
```

```python
# Araç renkleri oranı;

# Stil olarak "ggplot" kullan
plt.style.use("ggplot")

# Şeklin boyutunu belirle
plt.figure(figsize=(16, 6))

# Çubuk grafik çizimi
plt.bar(top10color.index, top10color.values, color="deeppink")

# Grafiği görüntüle
plt.show()
```

```python
# 'Marka' sütununa göre gruplayıp 'Fiyat' sütununun ortalamasını alın
# Ardından ortalamaları küçükten büyüğe sıralayın
fiyat_ortalamalari = df.groupby('Marka').Fiyat.mean().sort_values(ascending=True)

# Ortalamaların ondalık kısmını kaldırın
fiyat_ortalamalari = fiyat_ortalamalari.astype(int)

# Sonucu gösterin
print(fiyat_ortalamalari)
```

```python
# 'Marka' sütununa göre gruplayıp 'Fiyat' sütununun ortalamasını alın
# Ardından ortalamaları büyükten küçüğe sıralayın ve çubuk grafik çizin
plt.figure(figsize=(20, 10))
# Y eksenine uygun etiketleri ekleyin
plt.yticks([10000,100000,200000,300000,400000,500000,600000,700000,800000,900000,1000000 ],
['10k', '100k', '200k', '300k', '400k', '500k', '600k', '700k', '800k', '900k', '1000k'])
df.groupby('Marka').Fiyat.mean().sort_values(ascending=False).plot(kind='bar',color="#00A7FF")
```

```python
# Tofaş otomobillerinin model yıllarına göre ortalama fiyatı;

# "Tofaş" markasına ait araçların model yıllarına göre ortalama fiyatını gösteren çubuk grafik
plt.figure(figsize=(20, 6))

# "Marka" sütunu "Tofaş" olan verileri filtreleyin ve "Model Yıl" ile "Fiyat" sütunlarını seçin
# Ardından bu verilere göre çubuk grafik çizin
sns.barplot(x=df.loc[df["Marka"] == "Tofaş"]["Model Yıl"], y=df["Fiyat"],color="#6C5B7B")

# Grafiği göster
plt.show()
```

```python
# Araç model yıllarının ortalaması;
round(df.groupby("Marka")[["Model Yıl"]].mean())
```

```python
# Satıcıya Göre Ortalama Fiyatlar;

# Grafiği çizim boyutunu belirleyin
plt.figure(figsize=(12, 4))

# sns.barplot ile çubuk grafiği çizin
sns.barplot(x=df["Kimden"], y=df["Fiyat"], data=df,color="#0098FF")
```

```python
# Araç durumuna Göre Fiyatlar;

# Grafiği çizim boyutunu belirleyin
plt.figure(figsize=(12, 4))

# sns.barplot ile çubuk grafiği çizin
sns.barplot(x=df["Durum"], y=df["Fiyat"], data=df,color="#0098FF")
```

```python
# Yakıt türleri ve miktarları
yakıtc = df['Yakıt Turu'].value_counts()

# Pasta grafiği çizimi
plt.pie(yakıtc.values, labels=yakıtc.index, autopct='%1.2f%%', explode=[0, 0.1, 0.2, 0, 1.3])

# Grafiği göster
plt.show()
```

```python
# Vites türleri oranı;

# 'Vites Türü' sütunundaki değer sayılarını içeren bir değişken oluşturun
vites_turu_sayilari = df['Vites'].value_counts()

# Pasta grafiği çizimi
plt.pie(vites_turu_sayilari.values, labels=vites_turu_sayilari.index, autopct="%1.2f%%", explode=[0.1, 0, 0.1])
plt.style.use("default")
plt.show()
```
